# 📊 LLM Sales Simulator - Análisis de Datos

Este notebook analiza los resultados de las simulaciones de conversaciones de ventas.

**Estrategias comparadas:**
- Estática: 5,000 conversaciones con enfoque tradicional
- Dinámica: 5,000 conversaciones con enfoque adaptativo

**Total:** 10,000+ conversaciones
**Métricas:** 15+ indicadores por conversación

## 1. Importar Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configurar estilos
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print('✅ Librerías cargadas correctamente')
```

## 2. Cargar Datos

```python
# Cargar archivo de conversaciones
df = pd.read_csv('data/results/conversaciones.csv')

print(f'📊 Datos cargados: {len(df)} conversaciones')
print(f'📋 Columnas: {df.shape[1]}')
print(f'\n⏰ Primeras filas:')
df.head()
```

## 3. Exploración General de Datos

```python
# Info general
print('📊 INFORMACIÓN GENERAL:')
print(f'Total conversaciones: {len(df)}')
print(f'Total columnas: {df.shape[1]}')
print(f'\n🎯 Resultados:')
print(df['resultado_conversacion'].value_counts())
print(f'\n📈 Distribución %:')
print(df['resultado_conversacion'].value_counts(normalize=True) * 100)

# Estadísticas descriptivas
print(f'\n📐 Estadísticas básicas:')
df.describe()
```

## 4. Comparación de Estrategias (Estática vs Dinámica)

```python
# Separar por estrategia
if 'estrategia' in df.columns:
    estatico = df[df['estrategia'] == 'Estático']
    dinamico = df[df['estrategia'] == 'Dinámico']
else:
    # Si no existe columna estrategia, intentar inferir del nombre del vendedor
    print('⚠️ Nota: Columna "estrategia" no encontrada')
    print('Asume que tu DataFrame tiene esta columna')
    estatico = None
    dinamico = None

if estatico is not None and dinamico is not None:
    print('🔍 COMPARACIÓN DE ESTRATEGIAS:\n')
    
    # Tasas de éxito
    tasa_exito_estatico = (estatico['resultado_conversacion'] == 'EXITO').sum() / len(estatico) * 100
    tasa_exito_dinamico = (dinamico['resultado_conversacion'] == 'EXITO').sum() / len(dinamico) * 100
    
    print(f'📊 Tasa de ÉXITO:')
    print(f'  Estrategia Estática: {tasa_exito_estatico:.2f}%')
    print(f'  Estrategia Dinámica: {tasa_exito_dinamico:.2f}%')
    print(f'  Diferencia: {tasa_exito_dinamico - tasa_exito_estatico:+.2f}%')
    
    # Sentimiento promedio
    print(f'\n💭 Sentimiento Cliente (promedio):')
    print(f'  Estrategia Estática: {estatico["sentimiento_promedio_cliente"].mean():.3f}')
    print(f'  Estrategia Dinámica: {dinamico["sentimiento_promedio_cliente"].mean():.3f}')
    
    # Tasa de objeción
    print(f'\n❌ Tasa de Objeción Cliente (promedio):')
    print(f'  Estrategia Estática: {estatico["tasa_objecion_cliente"].mean():.2f}%')
    print(f'  Estrategia Dinámica: {dinamico["tasa_objecion_cliente"].mean():.2f}%')
```

## 5. Visualización de Resultados

```python
# Gráfico 1: Distribución de resultados
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico izquierdo: Conteo
resultado_counts = df['resultado_conversacion'].value_counts()
colors = ['#2ecc71', '#e74c3c', '#95a5a6']  # Verde, Rojo, Gris
axes[0].bar(resultado_counts.index, resultado_counts.values, color=colors[:len(resultado_counts)])
axes[0].set_title('Distribución de Resultados', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Número de Conversaciones')
axes[0].grid(axis='y', alpha=0.3)

# Gráfico derecho: Porcentaje
resultado_pct = df['resultado_conversacion'].value_counts(normalize=True) * 100
axes[1].pie(resultado_pct.values, labels=resultado_pct.index, autopct='%1.1f%%',
            colors=colors[:len(resultado_pct)], startangle=90)
axes[1].set_title('Porcentaje de Resultados', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print('✅ Gráfico 1: Distribución de resultados')
```

```python
# Gráfico 2: Comparación de estrategias (si existen)
if estatico is not None and dinamico is not None:
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Tasa de éxito
    estrategias = ['Estática', 'Dinámica']
    tasas_exito = [
        (estatico['resultado_conversacion'] == 'EXITO').sum() / len(estatico) * 100,
        (dinamico['resultado_conversacion'] == 'EXITO').sum() / len(dinamico) * 100
    ]
    axes[0, 0].bar(estrategias, tasas_exito, color=['#3498db', '#e74c3c'])
    axes[0, 0].set_title('Tasa de Éxito por Estrategia', fontweight='bold')
    axes[0, 0].set_ylabel('Porcentaje (%)')
    axes[0, 0].grid(axis='y', alpha=0.3)
    
    # Sentimiento del cliente
    sentimientos = [
        estatico['sentimiento_promedio_cliente'].mean(),
        dinamico['sentimiento_promedio_cliente'].mean()
    ]
    axes[0, 1].bar(estrategias, sentimientos, color=['#3498db', '#e74c3c'])
    axes[0, 1].set_title('Sentimiento Promedio del Cliente', fontweight='bold')
    axes[0, 1].set_ylabel('Sentimiento (-1 a 1)')
    axes[0, 1].axhline(y=0, color='black', linestyle='--', alpha=0.3)
    axes[0, 1].grid(axis='y', alpha=0.3)
    
    # Tasa de objeción
    objeciones = [
        estatico['tasa_objecion_cliente'].mean(),
        dinamico['tasa_objecion_cliente'].mean()
    ]
    axes[1, 0].bar(estrategias, objeciones, color=['#3498db', '#e74c3c'])
    axes[1, 0].set_title('Tasa de Objeción del Cliente', fontweight='bold')
    axes[1, 0].set_ylabel('Porcentaje (%)')
    axes[1, 0].grid(axis='y', alpha=0.3)
    
    # Preguntas abiertas del vendedor
    preguntas = [
        estatico['tasa_preguntas_abiertas_vendedor'].mean(),
        dinamico['tasa_preguntas_abiertas_vendedor'].mean()
    ]
    axes[1, 1].bar(estrategias, preguntas, color=['#3498db', '#e74c3c'])
    axes[1, 1].set_title('Preguntas Abiertas del Vendedor', fontweight='bold')
    axes[1, 1].set_ylabel('Porcentaje (%)')
    axes[1, 1].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print('✅ Gráfico 2: Comparación de estrategias')
```

## 6. Análisis de Vendedores

```python
# Rendimiento por vendedor
print('🎯 RENDIMIENTO POR VENDEDOR:\n')

vendedor_stats = df.groupby('vendedor_nombre').agg({
    'id_conversacion': 'count',
    'resultado_conversacion': lambda x: (x == 'EXITO').sum() / len(x) * 100,
    'sentimiento_promedio_cliente': 'mean',
    'tasa_objecion_cliente': 'mean'
}).rename(columns={
    'id_conversacion': 'Conversaciones',
    'resultado_conversacion': 'Tasa Éxito (%)',
    'sentimiento_promedio_cliente': 'Sentimiento Promedio',
    'tasa_objecion_cliente': 'Tasa Objeción Promedio'
})

vendedor_stats = vendedor_stats.sort_values('Tasa Éxito (%)', ascending=False)
print(vendedor_stats.round(2))

# Top 3 vendedores
print(f'\n🏆 Top 3 Vendedores por Tasa de Éxito:')
for i, (vendedor, row) in enumerate(vendedor_stats.head(3).iterrows(), 1):
    print(f'{i}. {vendedor}: {row["Tasa Éxito (%)"]:.2f}%')
```

```python
# Gráfico: Rendimiento vendedores
fig, ax = plt.subplots(figsize=(12, 6))

vendedor_exito = df.groupby('vendedor_nombre').apply(
    lambda x: (x['resultado_conversacion'] == 'EXITO').sum() / len(x) * 100
).sort_values(ascending=False)

colors_vendedores = plt.cm.RdYlGn(np.linspace(0.2, 0.8, len(vendedor_exito)))
ax.barh(vendedor_exito.index, vendedor_exito.values, color=colors_vendedores)
ax.set_xlabel('Tasa de Éxito (%)', fontweight='bold')
ax.set_title('Tasa de Éxito por Vendedor', fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print('✅ Gráfico: Rendimiento por vendedor')
```

## 7. Análisis de Clientes

```python
# Rendimiento por cliente
print('👥 RENDIMIENTO POR CLIENTE:\n')

cliente_stats = df.groupby('cliente_nombre').agg({
    'id_conversacion': 'count',
    'resultado_conversacion': lambda x: (x == 'EXITO').sum() / len(x) * 100,
    'tasa_objecion_cliente': 'mean',
    'tasa_preguntas_relevantes_cliente': 'mean'
}).rename(columns={
    'id_conversacion': 'Conversaciones',
    'resultado_conversacion': 'Tasa Éxito (%)',
    'tasa_objecion_cliente': 'Objeciones Promedio',
    'tasa_preguntas_relevantes_cliente': 'Preguntas Relevantes (%)'
})

cliente_stats = cliente_stats.sort_values('Tasa Éxito (%)', ascending=False)
print(cliente_stats.round(2))

# Clientes más difíciles
print(f'\n😤 Clientes más difíciles (menor tasa de compra):')
for i, (cliente, row) in enumerate(cliente_stats.tail(3).iterrows(), 1):
    print(f'{i}. {cliente}: {row["Tasa Éxito (%)"]:.2f}%')
```

## 8. Análisis de Correlaciones

```python
# Crear variable binaria para éxito
df['exito_binario'] = (df['resultado_conversacion'] == 'EXITO').astype(int)

# Columnas numéricas para correlación
numeric_cols = [
    'sentimiento_promedio_cliente',
    'tasa_objecion_cliente',
    'tasa_preguntas_relevantes_cliente',
    'tasa_palabras_dudosas_cliente',
    'frecuencia_preocupacion_principal_cliente',
    'tasa_preguntas_abiertas_vendedor',
    'tasa_interrupciones_vendedor',
    'polaridad_sentimiento_vendedor',
    'exito_binario'
]

# Filtrar columnas que existen
numeric_cols = [col for col in numeric_cols if col in df.columns]

# Calcular correlación
correlacion = df[numeric_cols].corr()['exito_binario'].sort_values(ascending=False)

print('📊 CORRELACIÓN CON ÉXITO\n')
print('Factores que más correlacionan con ÉXITO:')
for metrica, valor in correlacion.head(6).items():
    print(f'  {metrica:45s}: {valor:+.3f}')

print('\nFactores que menos correlacionan (o correlacionan negativamente):')
for metrica, valor in correlacion.tail(3).items():
    print(f'  {metrica:45s}: {valor:+.3f}')
```

```python
# Visualizar correlaciones
fig, ax = plt.subplots(figsize=(10, 6))

# Excluir 'exito_binario' de la visualización
corr_sin_exito = correlacion.drop('exito_binario').sort_values()

colors = ['#e74c3c' if x < 0 else '#2ecc71' for x in corr_sin_exito.values]
ax.barh(range(len(corr_sin_exito)), corr_sin_exito.values, color=colors)
ax.set_yticks(range(len(corr_sin_exito)))
ax.set_yticklabels(corr_sin_exito.index, fontsize=9)
ax.set_xlabel('Correlación con Éxito', fontweight='bold')
ax.set_title('Factores que Correlacionan con Conversaciones Exitosas', fontsize=14, fontweight='bold')
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print('✅ Gráfico: Correlaciones con éxito')
```

## 9. Conclusiones y Recomendaciones

```python
print('='*70)
print('RESUMEN EJECUTIVO - ANÁLISIS DE SIMULACIONES')
print('='*70)

print(f'\n📊 ESTADÍSTICAS GENERALES:')
print(f'  • Total conversaciones: {len(df):,}')
print(f'  • Éxitos: {(df["resultado_conversacion"] == "EXITO").sum():,} ({(df["resultado_conversacion"] == "EXITO").sum()/len(df)*100:.1f}%)')
print(f'  • Fallos: {(df["resultado_conversacion"] == "FALLO").sum():,} ({(df["resultado_conversacion"] == "FALLO").sum()/len(df)*100:.1f}%)')
print(f'  • Neutros: {(df["resultado_conversacion"] == "NEUTRO").sum():,} ({(df["resultado_conversacion"] == "NEUTRO"].sum()/len(df)*100:.1f}%)')

if estatico is not None and dinamico is not None:
    tasa_exito_estatico = (estatico['resultado_conversacion'] == 'EXITO').sum() / len(estatico) * 100
    tasa_exito_dinamico = (dinamico['resultado_conversacion'] == 'EXITO').sum() / len(dinamico) * 100
    
    print(f'\n🎯 COMPARACIÓN DE ESTRATEGIAS:')
    print(f'  • Estrategia Estática: {tasa_exito_estatico:.1f}% éxito')
    print(f'  • Estrategia Dinámica: {tasa_exito_dinamico:.1f}% éxito')
    
    mejora = tasa_exito_dinamico - tasa_exito_estatico
    if mejora > 0:
        print(f'  ✅ La estrategia dinámica es {mejora:.1f}% más efectiva')
    else:
        print(f'  ⚠️  La estrategia estática es {abs(mejora):.1f}% más efectiva')

print(f'\n🏆 TOP INSIGHTS:')
top_factor = correlacion.drop('exito_binario').idxmax()
print(f'  • Factor más importante: {top_factor} (r={correlacion[top_factor]:.3f})')
print(f'  • Vendedor más efectivo: {vendedor_exito.index[0]}')
print(f'  • Cliente más fácil de convertir: {cliente_stats.index[0]}')

print(f'\n💡 RECOMENDACIONES:')
print(f'  1. Enfocarse en {top_factor}')
print(f'  2. Entrenar vendedores con técnicas de {cliente_stats.index[0] if tasa_exito_dinamico > tasa_exito_estatico else "estáticas"}')
print(f'  3. Desarrollar estrategias personalizadas por tipo de cliente')
print(f'  4. Analizar técnicas del vendedor más exitoso')

print('\n' + '='*70)
```

## 10. Exportar Datos para Análisis Posterior

```python
# Exportar estadísticas por vendedor
vendedor_stats.to_csv('data/results/analisis_vendedores.csv')
print('✅ Estadísticas de vendedores exportadas a: analisis_vendedores.csv')

# Exportar estadísticas por cliente
cliente_stats.to_csv('data/results/analisis_clientes.csv')
print('✅ Estadísticas de clientes exportadas a: analisis_clientes.csv')

# Exportar correlaciones
correlacion.to_csv('data/results/correlaciones.csv')
print('✅ Correlaciones exportadas a: correlaciones.csv')

print('\n✅ Análisis completado')
```
